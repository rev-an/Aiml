```python
bc['diagnosis']= bc['diagnosis'].map({'B':0 , 'M':1})
```

```python
x= bc.drop('diagnosis', axis=1)
```

```python
y=bc['diagnosis']
```

```python
y
```

```python
scaler= StandardScaler()
x_scaled= scaler.fit_transform(x)
```

```python
x_scaled
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_scaled, y, test_size=0.3, random_state=42)
```

```python
svm_linear = SVC(kernel='linear', C=1)
svm_linear.fit(x_train, y_train)
```

```python
svm_rbf = SVC(kernel='rbf', C=1, gamma='scale')
svm_rbf.fit(x_train, y_train)
```

```python
feature_pair = ['radius_mean', 'texture_mean']
x_vis = bc[feature_pair].values
x_vis_scaled = scaler.fit_transform(x_vis)
```

```python
svm_vis = SVC(kernel='rbf', C=1, gamma='scale')
svm_vis.fit(x_vis_scaled, y)
```

```python
def plot_decision_boundary_2D(x, y, model, title, feature_names):
    h = .02
    x_min, x_max = x[:, 0].min() - 1, x[:, 0].max() + 1
    y_min, y_max = x[:, 1].min() - 1, x[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.astype(float)
    Z = Z.reshape(xx.shape)

    plt.figure(figsize=(8, 6))
    plt.contourf(xx, yy, Z, cmap=plt.cm.coolwarm, alpha=0.8)
    plt.scatter(x[:, 0], x[:, 1], c=y, cmap=plt.cm.coolwarm, edgecolors='k')
    plt.xlabel(feature_names[0])
    plt.ylabel(feature_names[1])
    plt.title(title)
    plt.show()
```

```python
plot_decision_boundary_2D(
    x_vis_scaled, y.values,
    svm_vis,
    "SVM Decision Boundary (RBF Kernel, 2 Features)",
    feature_names=feature_pair
)
```

```python
param_grid = {
    'C': [0.1, 1, 10, 100],
    'gamma': ['scale', 0.01, 0.1, 1]
}
grid = GridSearchCV(SVC(kernel='rbf'), param_grid, cv=5)
```

```python
grid.fit(x_scaled, y)
```

```python
print("🔧 Best Parameters:", grid.best_params_)
print("✅ Best CV Score:", grid.best_score_)
```

```python
best_model = grid.best_estimator_
cv_scores = cross_val_score(best_model, x_scaled, y, cv=5)
```

```python
y_pred = best_model.predict(x_test)
print("Classification Report (Test Set):")
print(classification_report(y_test, y_pred))
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```
