```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
numerics= td.select_dtypes(include= ['number']).columns
```

```python
for col in numerics:
    plt.figure(figsize=(6, 4))
    sns.histplot(td[col], kde=True, bins=30)
    plt.title(f'Histogram of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.tight_layout()
    plt.show()
```

```python
for col in numerics:
    plt.figure(figsize=(6, 4))
    sns.boxplot(x=td[col], color='lightcoral')
    plt.title(f'Boxplot of {col}')
    plt.xlabel(col)
    plt.tight_layout()
    plt.show()
```

```python
numerics_2= td.select_dtypes(include=['number'])
plt.figure(figsize=(10, 8))
corr_matrix = numerics_2.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', square=True)
plt.title('Correlation Matrix')
plt.tight_layout()
plt.show()
```
